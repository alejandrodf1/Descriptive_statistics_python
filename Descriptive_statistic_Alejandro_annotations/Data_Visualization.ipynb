# Data Visualization

For this class we are going to analyze the scatter plot. However, this class use a new data set called iris.

```
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')

iris.head()
```

```
#scatterplot
sns.scatterplot(data=iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

The following graph help in visualization

```
# join plot
sns.jointplot(data=iris, x = 'sepal_length', y = 'petal_length', hue='species')
```

```
#boxplot
sns.boxplot(x = 'species', y='sepal_length', data = iris)
```

```
#Barplot
mean_length = iris.groupby('species')['sepal_length'].mean()
print(mean_length)
sns.barplot(data = iris, x= 'species' , y='sepal_length')
```
